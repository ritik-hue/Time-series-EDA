```python

from pandas_datareader import data as pdr
import pandas as pd
from datetime import datetime
import yfinance as yf
```

```python
yf.pdr_override()
df1=pdr.get_data_yahoo('TSLA')
df1
```

```python
##sma
df1['Open:10 days rolling']=df1['Open'].rolling(window=10,min_periods=1).mean()
```

```python
df1[['Open','Open:10 days rolling']].plot(xlim=['2020-01-01','2021-01-01'])
```

```python
##cma
df1['Open'].expanding().mean().plot(figsize=(12,6))
```

```python
df1['EMA_0.1']=df1['Open'].ewm(alpha=0.1,adjust=False).mean()
```

```python
df1[['Open','EMA_0.1']].plot(xlim=['2020-01-01','2021-01-01'])
```
